### Star Cancellation Feature Extraction

```python
# Takes targets hit and missed per and returns the error percentage
def omission_error_per(targets_hit, targets_missed) :
    total_targets = targets_hit + targets_missed
    error_per = (targets_missed / total_targets) * 100
    
    return error_per

test_error = omission_error_per(75, 25)

print(test_error)
```
